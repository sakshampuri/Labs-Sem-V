### 5. Creating a circular mask on the input image. 
Hint: Slicing and masking with numpy arrays can be used to create a circular mask on the input image.


```python
lena = np.array(Image.open('data/sample.jpg'))

lx, ly, _ = lena.shape

X, Y = np.ogrid[0:lx, 0:ly]

mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > (lx * ly / 4)

lena[mask,:] = 0

Image.fromarray(lena)
```
